```python
#Importar Bibliotecas
import pandas as pd
```

#### Carregando arquivos FieldPRO

Sensor_FieldPRO, dados transmitidos pelo sensor

```python
# Carregar arquivo de dados transmitidos pelo sensor
name_file = 'Sensor_FieldPRO.csv'
diretorio = '../datasets/{name_file}'

df_sensor = pd.read_csv(diretorio.format(name_file=name_file), sep=',', decimal='.' )
df_sensor.head()
```

```python
df_sensor.info()
```

* Data e hora estão campo time-stamp ISO-8601 cujo time zone é UTC-0, 
    * contudo não detectado Dtype correto pelo pandas, indicando alguma linha com dado incorreto
* Observado valores nullos para humidade do ar e temperatura

___________________________________________________
Corrigindo formato time-stamp do campo data

```python
df_sensor['datetime'] = pd.to_datetime(df_sensor['Datetime – utc'], format='ISO8601')
df_sensor.head()
```

```python
df_sensor.info()
```

```python
df_sensor.describe()
```

_____________________________________________________________________________________________________________________________

#### Carregando arquivos Estação Convencional

Estacao_Convencional, dados transmitidos pela estão metereológica próxima

```python
name_file = 'Estacao_convencional.csv'
diretorio = '../datasets/{name_file}'

df_estacao = pd.read_csv(diretorio.format(name_file=name_file), sep=',', decimal='.' )
df_estacao.head()
```

```python
df_estacao.info()
```

```python
df_estacao.describe()
```

* Nenhum valor nulo observado
* Observados 
    * 2256 registros transmitidos pela estação metereológica
    * 1711 registros transmitidos pela placa FieldPro


* Campos data e hora colunas distintas, e time-zone UTC-3h00 (Brasília).
    * Observar horário de verão para o período: _verificado ano de 2020 não houve horário de verão_
    * Necessário compatibilizar bases de tempo(data/hora)

```python
df_estacao['datetime'] = pd.to_datetime(df_estacao['data'] + ' ' + df_estacao['Hora (Brasília)'] + '-03:00', format='ISO8601')
df_estacao.head()
```

```python
df_estacao.info()
```

________

* Observando período de transmissão dos dados

```python
df_estacao['datetime'].min()
```

```python
df_estacao['datetime'].max()
```

```python
df_sensor['datetime'].min()
```

```python
df_sensor['datetime'].max()
```

* Período (_não há horário de verão para o ano de 2020_):
    
    * Estação metereológica: 0h00 01/09/2020 a 23h00 03/12/2020 (UTC+3)
    * Sensor FieldPRO: 23h00 30/09/2020 a 13h00 11/12/2020 (UTC)


Realizar merge nos dados compatibilizando data/hora considerando UTC
* DE 23h00 30/09/2020 UTC
* ATÉ 23h00 03/12/2020 UTC-3

```python
df_merged = pd.merge(df_sensor, df_estacao, on='datetime')
df_merged.head()
```

Realizado conferência colunas "Datetime - utc",  "Data" e "Hora (Brasília)" estão compatíveis
* 2020-09-30T23:00:00Z equivale a 2020-09-03 20:00 (horário de Brasília)

```python
#Selecionando colunas relevantes
df_merged = df_merged[['datetime', 'air_humidity_100', 'air_temperature_100','atm_pressure_main','num_of_resets','piezo_charge','piezo_temperature','chuva']]
df_merged.head()
```

```python
df_merged.info()
```

```python
#Verificando registros com valores nulos
df_merged.loc[df_merged['air_humidity_100'].isnull()]
```

Observado:
* Registros nulos de "air_humidity_100" ocorrem ao mesmo tempo quando nulos "air_temperature_100".
* Os registros são nulos sem ocorrência de chuva.

Em virtude do observado optou-se por remover linhas com dados nulos.

```python
df_merged.dropna(subset=['air_humidity_100'], inplace=True)
```

```python
df_merged.info()
```

```python
df_merged.describe()
```

# Exportando dataset tratado e combinado.

```python
df_merged.to_csv('../datasets/df_merged.csv', sep = ';', index=False)
```
